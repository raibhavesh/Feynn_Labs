```python
import pandas as pd
import numpy as np

MD_x = mcdonalds.iloc[:, 0:11]


MD_x = (MD_x == "Yes").astype(int)

col_means = MD_x.mean().round(2)


print(col_means)
```

```python
from sklearn.decomposition import PCA
MD_pca = PCA()

MD_pca.fit(MD_x)
explained_variance_ratio = MD_pca.explained_variance_ratio_
singular_values = MD_pca.singular_values_
components = MD_pca.components_


print("Explained Variance Ratio:")
print(explained_variance_ratio)


print("\nSingular Values:")
print(singular_values)


print("\nPrincipal Components:")
print(components)
```

```python

print("PCA Results:")
print("Standard deviations (Singular Values):")
print(np.round(MD_pca.singular_values_, 1))

print("\nProportion of Variance Explained:")
print(np.round(MD_pca.explained_variance_ratio_, 1))

print("\nPrincipal Components:")
print(np.round(MD_pca.components_, 1))
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


# Plot the PCA results
plt.figure(figsize=(10, 5))


projection = MD_pca.transform(MD_x)
plt.scatter(projection[:, 0], projection[:, 1], c='grey', alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA Projection')

# Plot the principal axes
plt.figure(figsize=(5, 5))
for i, (comp, var) in enumerate(zip(MD_pca.components_, MD_pca.explained_variance_ratio_)):
    plt.arrow(0, 0, comp[0], comp[1], head_width=0.2, head_length=0.2, linewidth=2, color='red')
    plt.text(comp[0] + 0.05, comp[1] + 0.05, f'PC{i + 1} ({var:.2f})', color='red')

plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Principal Components (Red Arrows)')

plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import random

random.seed(1234)
np.random.seed(1234)


k_values = range(2, 9)

best_k = None
best_kmeans = None
best_silhouette_score = -1

for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
    kmeans.fit(MD_x)


    silhouette_avg = silhouette_score(MD_x, kmeans.labels_)

    if silhouette_avg > best_silhouette_score:
        best_k = k
        best_kmeans = kmeans
        best_silhouette_score = silhouette_avg


print(f"Best number of clusters (k): {best_k}")
print(f"Best silhouette score: {best_silhouette_score}")

relabeling_mapping = {label: i for i, label in enumerate(sorted(set(best_kmeans.labels_)))}
best_kmeans.labels_ = [relabeling_mapping[label] for label in best_kmeans.labels_]
```
